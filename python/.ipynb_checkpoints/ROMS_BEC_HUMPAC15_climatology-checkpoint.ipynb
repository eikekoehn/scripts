```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
import xarray as xr
import cmocean as cm
```

```python
# choose the variable to plot the climatology for
varstring = 'O2'
```

```python
# other settings
plt.rcParams['figure.figsize'] = [15, 6]
```

```python
#open datasets
ds = xr.open_dataset('/net/kryo/work2/fana/hindcast_bec/analysis/z_humpac15_avg_1979-2016_clim.nc')
print('Data Opened')
grid = xr.open_dataset('/net/nardus/work/Ana/masks_and_maps/humpac_grids/humpac15_sm5rf015_grd.nc')
print('Grid Opened')
```

```python
# choose location
eta = 100
xi = 100
```

```python
# choose dataset and read it in
data_cont = eval('ds.'+varstring+'[:,:,eta,xi]')
units = eval('ds.'+varstring+'.units')
print('Data read in!')

# read in coordinates and units
lati = grid.lat_rho.values[eta,xi].round(decimals=2)
loni = grid.lon_rho.values[eta,xi].round(decimals=2)
depth_coor = ds.depth

# create month axis
from datetime import datetime
monax = [datetime(2017, _, 1) for _ in range(1,13)]
```

```python
# temporal average of data
data_plot = np.nanmean(data_cont,axis=0)
```

```python
# choose the colormap
if varstring == 'v':
    colmap = cm.cm.delta
elif varstring == 'u':
    colmap = cm.cm.delta
elif varstring == 'O2':
    colmap = cm.cm.oxy
elif varstring == 'temp':
    colmap = cm.cm.thermal
elif varstring == 'salt':
    colmap = cm.cm.haline
else:
    colmap = cm.cm.deep
```

```python
plt.close()
fig,(ax0,ax1) = plt.subplots(1,2)
h0 = ax0.contourf(monax,depth_coor,data_cont.T,levels=14,cmap=colmap)
ax0.set_title(varstring+' climatology at '+str(lati)+'N and '+str(loni)+'E')
ax0.set_ylabel('Depth in m')
ax0.set_xlabel('Month')
ax0.set_xticklabels([datetime(2017,_,1).strftime('%b') for _ in range(1,13)])
plt.colorbar(h0,ax=ax0)
ax1.plot(data_plot,depth_coor)
ax1.set_title('Annual Mean '+varstring+' at '+str(lati)+'N and '+str(loni)+'E')
ax1.set_xlabel(units)
ax1.grid()
plt.show()
```
